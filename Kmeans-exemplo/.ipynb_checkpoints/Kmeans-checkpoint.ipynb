## 1. Fazendo os imports:

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt  
from sklearn.cluster import KMeans 
```

## 2. Importando a base de dados:

```python
headers = ['sepal length', 'sepal width', 'petal length', 'petal width', 'class']
dataset = pd.read_csv("./iris.data", encoding = "ISO-8859-1", decimal=",", header=None, names=headers)
```

### 3. Convertendo os valores para o tipo float

```python
for col in  dataset.columns[0:4]:
    dataset[col] = dataset[col].astype(float)
```

### 3.1. Listando os tipos dos valores da base de dados:

```python
dataset.dtypes
```

## 4. Seprando as colunas referente as informações da coluna de classificação

```python
informacoes = dataset.iloc[:, 0:4]
classe = dataset.iloc[:, 4]
```

```python
dataset.head(5)
```

```python
informacoes.head(5)
```

```python
classe.head(5)
```

## 5. Aplicando o algoritmo K-MEANS:

```python
kmeans = KMeans(n_clusters=3).fit(informacoes)
informacoes_clustered = kmeans.fit_predict(informacoes)
```

```python
results = dataset[['class']].copy()
results['clusterNumber'] = informacoes_clustered
results
```

## 6. Salvando as informações da classe com seu respectivo cluster

```python
results.to_csv('clusters.csv', index=False, header=False, sep=';')
```

#### A nível de curiosidade, os centróides encontrados durante a execução do algoritmo:

```python
kmeans.cluster_centers_
```

```python
kmeans.cluster_centers_[:, 0]
```

```python

plt.scatter(informacoes.iloc[:, 0], informacoes.iloc[:,1], s = 100, c = kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'red', label = 'Centroids')
plt.title('Iris Clusters and Centroids')
plt.xlabel('SepalLength')
plt.ylabel('SepalWidth')
plt.legend()
#plt.savefig(title + '.jpg')
plt.show()
```

```python
plt.scatter(informacoes.iloc[:, 2], informacoes.iloc[:,3], s = 100, c = kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:, 2], kmeans.cluster_centers_[:, 3], s = 300, c = 'red', label = 'Centroids')
plt.title('Iris Clusters and Centroids')
plt.xlabel('PetalLength')
plt.ylabel('PetalWidth')
plt.legend()
#plt.savefig(title + '.jpg')
plt.show()
```
